```python
# Import dependencies
import pandas as pd
import numpy as np
import shutup; shutup.please()
from decimal import Decimal
from fractions import Fraction
```

```python
monthly_one = pd.read_csv("CleanedCSVs/StockIndexeswithMacroEconomic.csv")
monthly_one.head()
```

```python
monthly_one['Date'] = pd.to_datetime(monthly_one['Date'])
monthly_one.set_index('Date', inplace=True)
```

```python
df_monthly = monthly_one.resample('M').mean()
```

```python
df_monthly.head()
```

```python
df_monthly.drop(["Unnamed: 0"], axis=1)
```

```python
cpi_df = pd.read_csv("csv/CPIAUCSL_Monthly.csv")
cpi_df.head()
```

```python
cpi_df['Date'] = pd.to_datetime(cpi_df['Date'])
cpi_df.set_index('Date', inplace=True)
```

```python
merged_1 = pd.merge(df_monthly, cpi_df, on="Date", how="left")
```

```python
unrate_df = pd.read_csv("csv/UNRATE_Monthly.csv")
unrate_df.head()
```

```python
unrate_df['Date'] = pd.to_datetime(unrate_df['Date'])
unrate_df.set_index('Date', inplace=True)
```

```python
monthly_totals = pd.merge(merged_1, unrate_df, on="Date", how="left")
```
