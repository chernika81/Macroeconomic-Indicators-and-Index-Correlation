## Extract the Data

```python
# Read the data into a Pandas DataFrame
cpi_df = pd.read_csv('csv/CPIAUCSL_Monthly.csv')
dow_df = pd.read_csv('csv/Dow_Jones_daily.csv')
fedfunds_df = pd.read_csv('csv/FedFundsDaily.csv')
inflation_df = pd.read_csv('csv/InflationRateDaily.csv')
nasdaq_df = pd.read_csv('csv/Nasdaq_daily.csv')
sp_df = pd.read_csv('csv/S&P_daily.csv')
unrate_df = pd.read_csv('csv/UNRATE_Monthly.csv')
```

```python
merged_df_1 = pd.merge(fedfunds_df, dow_df, on="Date", how="left")
merged_df_2 = pd.merge(merged_df_1, inflation_df, on="Date", how="left")
merged_df_3 = pd.merge(merged_df_2, sp_df, on="Date", how="left")
merged_df_4 = pd.merge(merged_df_3, nasdaq_df, on="Date", how="left")
merged_df_5 = merged_df_4.dropna()
merged_df_5
```

```python
merged_df_6 = pd.merge(merged_df_5, cpi_df, on="Date", how="left")
merged_df = pd.merge(merged_df_6, unrate_df, on="Date", how="left")
merged_df
```
